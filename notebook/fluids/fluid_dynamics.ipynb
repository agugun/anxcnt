# 2D Fluid Dynamics - Finite Element Method (FEM)

This notebook demonstrates the numerical simulation of **2D Incompressible Fluid Flow** using the Finite Element Method.

## 1. Governing Equation (Stokes)

For steady-state incompressible flow at low Reynolds numbers:

**Momentum Balance**:
$$-\mu \nabla^2 \mathbf{u} + \nabla p = \mathbf{f}$$

**Continuity**:
$$\nabla \cdot \mathbf{u} = 0$$

## 2. Detailed Analytical Trace: Variational Form (Sympy)

We use Sympy to represent the weak form integration by parts for the Laplacian term.

```python
import sympy as sp
from IPython.display import display, Math

sp.init_printing()
mu = sp.symbols('mu', real=True, positive=True)
u = sp.Function('u')(x, y)
v = sp.Function('v')(x, y) # Test function
x, y = sp.symbols('x y')

# Laplacian integrated by parts: Integral(mu * grad(u) . grad(v))
laplacian_weak = mu * (u.diff(x)*v.diff(x) + u.diff(y)*v.diff(y))
display(Math(f"\\text{{Weak Form (Diffusion Term): }} \\int_{\\Omega} {sp.latex(laplacian_weak)} d\\Omega"))

# Pressure term: -Integral(p * div(v))
p = sp.Function('p')(x, y)
pressure_weak = -p * (v.diff(x) + v.diff(y))
display(Math(f"\\text{{Weak Form (Pressure Term): }} \\int_{\\Omega} {sp.latex(pressure_weak)} d\\Omega"))
```

## 3. Stabilization and Numerical Solution

The **axcnt_cpp** engine implements **PSPG (Pressure-Stabilized Petrov-Galerkin)** to solve these equations using P1-P1 elements. This allows for stable pressure recovery even with linear velocity/pressure interpolations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Lid-driven cavity conceptual flow-field
x_vals = np.linspace(0, 1, 20)
y_vals = np.linspace(0, 1, 20)
X, Y = np.meshgrid(x_vals, y_vals)
U = np.zeros_like(X)
V = np.zeros_like(Y)
U[-1, :] = 1.0 # Moving lid

plt.figure(figsize=(8, 8))
plt.quiver(X, Y, U, V)
plt.title("Analytical Trace: 2D Stokes Velocity Field (Cavity)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
