# 2D Dual-Porosity Reservoir Simulation (Warren-Root)

This notebook demonstrates the modeling of **Naturally Fractured Reservoirs (NFRs)** using the dual-porosity approach.

## 1. Governing Equations

The system is represented by two coupled PDEs (Matrix and Fracture):

**Fracture System**:
$$\nabla \cdot (k_f \nabla p_f) - \tau = \phi_f c_f \frac{\partial p_f}{\partial t}$$

**Matrix System**:
$$\tau = -\phi_m c_m \frac{\partial p_m}{\partial t}$$

Where $\tau = \sigma k_m \frac{p_m - p_f}{\mu}$ is the transfer term.

## 2. Detailed Analytical Trace (Sympy)

We use Sympy to represent the coupled system and discuss the linearization of the transfer term.

```python
import sympy as sp
from IPython.display import display, Math

sp.init_printing()
pf = sp.Function('p_f')(t)
pm = sp.Function('p_m')(t)
sigma, km, mu, phim, cm = sp.symbols('sigma k_m mu phi_m c_m', real=True, positive=True)

# Transfer Equation (simplified for a single block interaction)
transfer = sigma * (km / mu) * (pm - pf)
matrix_eq = sp.Eq(-phim * cm * pm.diff(t), transfer)

display(Math(f"\\text{{Matrix-Fracture Transfer: }} {sp.latex(matrix_eq)}"))

# Solving for the pressure lag in the matrix
matrix_sol = sp.dsolve(matrix_eq, pm)
display(Math(f"\\text{{Matrix Pressure Response: }} {sp.latex(matrix_sol)}"))
```
