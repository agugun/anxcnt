# Material Balance Analysis (MBA) - Under-saturated Oil Reservoirs

This notebook demonstrates the **Material Balance Equation (MBE)** and the **Havlena-Odeh** graphical method for estimating original oil in place (OOIP).

## 1. Governing Equation

For an under-saturated oil reservoir (no gas cap $m=0$, no water drive $W_e=0$):
$$F = N(E_o + E_{fw})$$

Where:
- $F$: Underground withdrawal
- $N$: Original Oil In Place (OOIP)
- $E_o$: Oil expansion factor
- $E_{fw}$: Formation and water expansion factor

## 2. Detailed Analytical Trace (Sympy)

We derive the expansion terms $E_o$ and $E_{fw}$ from fluid compressibility definitions.

```python
import sympy as sp
from IPython.display import display, Math

sp.init_printing()
Np, Bo, Boi, Rs, Rsi, Bg, Wp, Bw, Winj, Bwinj, N = sp.symbols('Np Bo Boi Rs Rsi Bg Wp Bw Winj Bwinj N', real=True)

# Total Withdrawal F
F = Np * (Bo + (Rsi - Rs) * Bg) + (Wp - Winj) * Bw
display(Math(f"\\text{{Withdrawal }} F = {sp.latex(F)}"))

# Total Expansion Et
Et = (Bo - Boi) + (Rsi - Rs) * Bg
display(Math(f"\\text{{Expansion }} E_t = {sp.latex(Et)}"))

# Havlena-Odeh Plot Relationship
ho_eq = sp.Eq(F, N * Et)
display(Math(f"\\text{{Havlena-Odeh Line: }} {sp.latex(ho_eq)}"))
```

## 3. Graphical Verification

In a plot of $F$ vs $E_t$, a straight line through the origin indicates a constant OOIP (N), where the slope is precisely the value of $N$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Mock Production Data
Et_data = np.linspace(0.01, 0.05, 10)
N_true = 50.0  # MMSTB
F_data = N_true * Et_data + np.random.normal(0, 0.02, 10)

plt.figure(figsize=(8, 6))
plt.scatter(Et_data, F_data, color='blue', label='Production Data')
plt.plot(Et_data, N_true * Et_data, 'r--', label=f'Havlena-Odeh Fit (N={N_true})')
plt.xlabel("Expansion factor Et")
plt.ylabel("Underground Withdrawal F")
plt.title("Material Balance: Havlena-Odeh Plot Trace")
plt.legend()
plt.grid(True)
plt.show()
```
