# 2D Oil-Gas Reservoir Simulation - Multi-phase Flow

This notebook demonstrates the mathematical modeling and numerical simulation of **Oil-Gas Multi-phase Flow** in Porous Media.

## 1. Governing Equation

The system is governed by conservation of mass for each phase (Oil and Gas):

**Oil Phase**:
$$\nabla \cdot \left( \frac{k k_{ro}}{\mu_o B_o} \nabla p_o \right) = \frac{\partial}{\partial t} \left( \frac{\phi S_o}{B_o} \right)$$

**Gas Phase** (including dissolved gas):
$$\nabla \cdot \left( \frac{k k_{rg}}{\mu_g B_g} \nabla p_g + R_s \frac{k k_{ro}}{\mu_o B_o} \nabla p_o \right) = \frac{\partial}{\partial t} \left( \frac{\phi S_g}{B_g} + \frac{\phi R_s S_o}{B_o} \right)$$

## 2. Detailed Analytical Trace (Sympy)

We use Sympy to represent the accumulation terms and discuss the coupling between pressure ($p$) and saturation ($S$).

```python
import sympy as sp
from IPython.display import display, Math

sp.init_printing()
t, phi, So, Sg, Bo, Bg, Rs = sp.symbols('t phi S_o S_g B_o B_g R_s', real=True)

# Accumulation Term for Oil
acc_oil = sp.diff(phi * So / Bo, t)
display(Math(f"\\text{{Oil Accumulation: }} {sp.latex(acc_oil)}"))

# Accumulation Term for Gas
acc_gas = sp.diff(phi * Sg / Bg + phi * Rs * So / Bo, t)
display(Math(f"\\text{{Gas Accumulation: }} {sp.latex(acc_gas)}"))

display(Math("\\text{The numerical solver must linearize these non-linear terms using a Newton-Raphson approach.}"))
```
