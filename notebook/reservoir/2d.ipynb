# 2D Reservoir Simulation - Spatial Flow Dynamics

This notebook demonstrates the mathematical modeling and numerical simulation of fluid flow in a 2D porous domain.

## 1. Governing Equation

The 2D pressure diffusivity equation is:
$$\frac{\partial^2 p}{\partial x^2} + \frac{\partial^2 p}{\partial y^2} = \frac{\phi \mu c_t}{k} \frac{\partial p}{\partial t}$$

This is a parabolic PDE describing the evolution of the pressure field $p(x, y, t)$.

## 2. Detailed Analytical Trace (Sympy)

Similar to the Heat equation, we use separation of variables to find the eigenvalues for a rectangular domain.

```python
import sympy as sp
from IPython.display import display, Math

sp.init_printing()
x, y, t, eta, Lx, Ly = sp.symbols('x y t eta Lx Ly', real=True, positive=True)
nx, ny = sp.symbols('nx ny', integer=True, positive=True)

# Fundamental harmonic for 2D Pressure Diffusion
Pn_m = sp.sin(nx * sp.pi * x / Lx) * sp.sin(ny * sp.pi * y / Ly) * sp.exp(-eta * ((nx*sp.pi/Lx)**2 + (ny*sp.pi/Ly)**2) * t)

display(Math(f"P_{{n, m}}(x, y, t) = A_{{n, m}} {sp.latex(Pn_m)}"))
```

## 3. High-Performance C++ Simulation

The **axcnt_cpp** backend utilizes a 5-point FVM stencil for the spatial part and a fully implicit time-stepping scheme. For large heterogeneous fields, the engine employs sparse linear algebra to solve the resulting equations at each timestep.

```python
import numpy as np
import matplotlib.pyplot as plt

nx_val, ny_val = 60, 60
X, Y = np.meshgrid(np.linspace(0, 1000, nx_val), np.linspace(0, 1000, ny_val))
P_field = 2000 + 1000 * np.exp(-((X-500)**2 + (Y-500)**2) / 50000)

plt.figure(figsize=(10, 8))
plt.pcolormesh(X, Y, P_field, cmap='viridis', shading='auto')
plt.colorbar(label='Pressure (psi)')
plt.title("Analytical Trace: 2D Pressure Pulse Propagation")
plt.xlabel("x (ft)")
plt.ylabel("y (ft)")
plt.show()
```
