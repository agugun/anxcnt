# 3D Reservoir Simulation - Full Field Dynamics

This notebook demonstrates the mathematical modeling and numerical simulation of fluid flow in a 3D porous domain.

## 1. Governing Equation

The 3D pressure diffusivity equation is:
$$\nabla \cdot \left( \frac{k}{\mu} \nabla p \right) = \phi c_t \frac{\partial p}{\partial t}$$

In Cartesian coordinates $(x, y, z)$:
$$\frac{\partial^2 p}{\partial x^2} + \frac{\partial^2 p}{\partial y^2} + \frac{\partial^2 p}{\partial z^2} = \frac{1}{\eta} \frac{\partial p}{\partial t}$$

## 2. Detailed Analytical Trace (Sympy)

We derive the 3D fundamental solution using separation of variables.

```python
import sympy as sp
from IPython.display import display, Math

sp.init_printing()
x, y, z, t, eta, Lx, Ly, Lz = sp.symbols('x y z t eta Lx Ly Lz', real=True, positive=True)
nx, ny, nz = sp.symbols('nx ny nz', integer=True, positive=True)

kx_n = nx * sp.pi / Lx
ky_m = ny * sp.pi / Ly
kz_l = nz * sp.pi / Lz

# 3D Harmonic Fundamental Solution
P_nml = sp.sin(kx_n*x) * sp.sin(ky_m*y) * sp.sin(kz_l*z) * sp.exp(-eta * (kx_n**2 + ky_m**2 + kz_l**2) * t)

display(Math(f"P_{{n, m, l}}(x, y, z, t) = A_{{n, m, l}} {sp.latex(P_nml)}"))
```
